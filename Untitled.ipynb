### One-hot Encoding  <a name="encoding"></a>

```python
def oneHot(column):
    arr = []
    for i in column:
        if(i in arr):
            continue
        else:
            arr.append(i)
    matrix = [[0 for x in range(len(arr))] for x in range(len(column))]
    j = 0
    for i in column:
        matrix[j][arr.index(i)] = 1
        j += 1
    df2 = pd.DataFrame(np.array(matrix),columns= arr)
    return df2
```
